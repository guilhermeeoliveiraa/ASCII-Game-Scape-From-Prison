- Sugestão de músicas com características semelhantes;

A base de dados utilizada contém um conjunto real e abrangente de músicas, com valores para diversos atributos acústicos, que representam diferentes aspectos sonoros e musicais de cada faixa. Esses atributos são:

- Danceability (Dançabilidade): mede o quão adequada é uma faixa para dançar, considerando ritmo, estabilidade e repetição;

- Energy (Energia): representa a intensidade e atividade da música, associada à percepção de energia;

- Key (Tonalidade): indica a tonalidade musical da faixa, correspondendo a uma das 12 notas da escala cromática;

- Loudness (Intensidade Sonora): volume geral da música, medido em decibéis;

- Mode (Modo): refere-se ao modo musical, geralmente maior ou menor, influenciando a sensação da música;

- Liveness (Vivacidade): indica a presença de público ao vivo na gravação;

- Valence (Valência): expressa a positividade ou humor da música, de alegre a melancólica;

- Time (Tempo): velocidade da música, medida em batidas por minuto (BPM).

Esses múltiplos atributos permitem representar cada música como um vetor multidimensional, tornando o problema adequado para técnicas de agrupamento que exploram estruturas complexas em espaços de alta dimensionalidade.

O método utilizado foi o K-Means Clustering, um algoritmo clássico de aprendizado não supervisionado, cuja finalidade é particionar os dados em k grupos distintos, minimizando a variabilidade interna dos grupos.

Funcionamento do K-Means:
1. Escolhe-se um valor k, que representa o número de grupos.

2. Inicialmente, k centróides são colocados aleatoriamente no espaço dos dados.

3. Cada ponto é atribuído ao centróide mais próximo, formando os grupos.

4. Os centróides são recalculados como a média dos pontos do grupo.

5. Os passos 3 e 4 se repetem até que os centróides não mudem mais (convergência).

Apesar de o conjunto de dados conter uma mistura de variáveis contínuas (como danceability, loudness, tempo) e variáveis discretas representadas numericamente (como key e mode), o K-Means ainda se mostra adequado como abordagem inicial pela sua simplicidade e capacidade de identificar estruturas de agrupamento baseadas em proximidade euclidiana. Como o foco está em explorar padrões ocultos sem rótulos prévios, K-Means oferece uma maneira eficiente de testar combinações de atributos e avaliar o quão bem esses agrupamentos se formam, especialmente ao utilizar métricas como o índice de silhueta. Além disso, sua interpretação é direta e seu desempenho é satisfatório em amostras moderadas, tornando-o apropriado para experimentação exploratória nesse tipo de análise.

Outros métodos como DBSCAN ou hierárquico poderiam ser considerados, mas K-Means é adequado quando se assume que os dados podem ser agrupados com base em distância euclidiana e número de grupos fixo.

Para verificar quais atributos influenciam mais na formação de agrupamentos bem definidos, utilizou-se o índice de silhueta, uma métrica que quantifica o quão bem um dado ponto está encaixado em seu cluster, comparado à proximidade com outros clusters. Valores próximos de 1 indicam agrupamentos bem separados e coesos, enquanto valores próximos de 0 indicam clusters sobrepostos.

Para isso, o K-Means foi aplicado em diferentes combinações de atributos, e o índice de silhueta foi calculado para cada resultado. Essa análise permite identificar quais características acústicas são mais relevantes para segmentar as músicas em grupos distintos, contribuindo para uma melhor compreensão da estrutura dos dados e para possíveis aplicações em sistemas de recomendação ou organização automática de músicas.

Outros métodos como DBSCAN ou clustering hierárquico poderiam ser considerados, mas o K-Means é particularmente adequado quando se assume que os dados podem ser agrupados com base em distância euclidiana e que o número de grupos é pré-determinado.

Comparado a outras bases de dados, este conjunto possui um equilíbrio entre quantidade de dados e qualidade das informações, facilitando a manipulação e o processamento computacional sem comprometer a representatividade musical. Além disso, sua estrutura já organizada e disponível publicamente permite uma integração rápida e eficiente no ambiente de desenvolvimento, reduzindo a complexidade de pré-processamento e problemas de inconsistência de dados. Assim, este dataset se mostrou mais apropriado para o objetivo de identificar agrupamentos naturais de músicas com base em características acústicas, possibilitando uma análise robusta e interpretável, alinhada ao escopo do estudo.

Embora o agrupamento não utilize rótulos explícitos, conhecer a diversidade de gêneros ajuda a orientar a escolha do número de clusters, garantindo que o modelo busque uma segmentação coerente com a variedade musical existente.
```python
# Verificar quantos gêneros únicos existem no dataset (antes da limpeza)
generos_unicos = df['genre'].unique()
numero_generos = len(generos_unicos)

print(f"Número total de gêneros únicos (antes da limpeza): {numero_generos}")
print("Gêneros encontrados (antes da limpeza):")
print(generos_unicos)
```

Durante a pesquisa esse código foi utilizado para verificar quais atributos entre os disponíveis no conjunto de dados de músicas são mais relevantes para a formação de clusters bem definidos (isto é, grupos de músicas com características semelhantes), utilizando o algoritmo KMeans de agrupamento não supervisionado. Para isso, é utilizado o índice de silhueta como métrica de avaliação da qualidade dos agrupamentos.

```python
# Lista de atributos a testar
columns = ['danceability', 'energy', 'key', 'loudness', 'mode', 'speechiness',
           'acousticness', 'instrumentalness', 'liveness', 'valence', 'tempo']

# Remove linhas com valores nulos nas colunas relevantes
df_clean = df[columns].dropna()

# Carregar os dados disponíveis
sample = df_clean.copy()

# Testa apenas combinações de dois atributos
results = []

for combo in combinations(columns, 2):  # Duplas
    try:
        X = sample[list(combo)]

        # Aplica KMeans com n clusters
        kmeans = KMeans(n_clusters=numero_generos, random_state=42, n_init=10)
        labels = kmeans.fit_predict(X)

        # Calcula o índice de silhueta
        score = silhouette_score(X, labels)
        results.append({'atributos': combo, 'silhouette': score})
    except Exception as e:
        print(f"Erro com atributos {combo}: {e}")

# Ordena os resultados do melhor para o pior
results = sorted(results, key=lambda x: x['silhouette'], reverse=True)

# Salva os top 10 em uma variável para uso posterior
top_combinations = [res['atributos'] for res in results[:10]]

# Mostra os top 10
print("\nTop 10 combinações com melhor índice de silhueta:")
for res in results[:10]:
    print(f"{res['atributos']} -> silhueta: {res['silhouette']:.4f}")
```

Os resultados obtidos indicam que o atributo 'mode' apresenta um índice de silhueta perfeito (1.0000) quando utilizado isoladamente para a formação de clusters, sugerindo que essa variável é extremamente eficaz para separar claramente os grupos no conjunto de dados analisado, isso se deve, principalmente, por esse ser um valor binário, mas que ainda é um bom metodo de classificação. Além disso, a combinação de 'mode' com 'instrumentalness' também alcança um valor muito alto (0.9851), o que reforça a importância desses dois atributos para a definição de agrupamentos bem distintos. O atributo 'instrumentalness' isoladamente também demonstra forte capacidade discriminativa (0.9771), corroborando seu papel relevante na segmentação das músicas. Combinações envolvendo 'mode' com 'speechiness', 'acousticness' e 'liveness' apresentam índices de silhueta acima de 0.79, indicando que, embora menos fortes que 'instrumentalness', essas variáveis contribuem de forma significativa para a formação de clusters coerentes. Outras combinações, como 'speechiness' com 'instrumentalness' e 'acousticness' com 'instrumentalness', assim como os atributos isolados 'liveness' e 'acousticness', obtêm valores moderados acima de 0.7, mostrando um efeito positivo na definição dos grupos, mas com menor clareza comparada aos primeiros resultados. Em suma, o experimento evidencia que os atributos relacionados à modalidade musical ('mode'), à presença ou ausência de instrumentos ('instrumentalness') e à características de presença vocal ('speechiness'), acústica e vivacidade ('acousticness' e 'liveness') são os principais responsáveis pela formação de clusters mais definidos no conjunto de dados estudado, sugerindo que aspectos qualitativos e texturais da música têm maior impacto na segmentação do que outros atributos analisados.

```python
# Garante que a combinação ('danceability', 'energy') esteja presente
if ('danceability', 'energy') not in top_combinations:
    top_combinations = [('danceability', 'energy')] + top_combinations

# Criar gráficos de cluster com 5 e com numero_generos clusters
for combo in top_combinations:
    X = sample[list(combo)]

    fig, axes = plt.subplots(1, 2, figsize=(14, 5))

    for ax, k in zip(axes, [5, numero_generos]):
        kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
        labels = kmeans.fit_predict(X)

        if len(combo) == 1:
            # Gráfico 1D com stripplot
            sns.stripplot(x=combo[0], y=[0]*len(X), hue=labels, data=X, palette='Set2', jitter=True, ax=ax)
            ax.set_yticks([])
            ax.set_ylabel('')
            ax.set_title(f'{combo[0]} - {k} Clusters')
        else:
            # Gráfico 2D com scatterplot
            sns.scatterplot(x=combo[0], y=combo[1], hue=labels, data=X, palette='Set2', ax=ax)
            ax.set_ylabel(combo[1])
            ax.set_title(f'{combo[0]} vs {combo[1]} - {k} Clusters')

        ax.set_xlabel(combo[0])
        ax.legend().remove()
        ax.grid(True)

    plt.tight_layout()
    plt.show()
```

Ao aplicar o algoritmo de agrupamento K-Means sobre diferentes combinações de atributos do conjunto de dados do Spotify, visualizamos os resultados por meio de gráficos bidimensionais, utilizando tanto uma quantidade fixa de 5 clusters quanto uma quantidade adaptada ao número de gêneros únicos identificados no dataset (neste caso, 59).

Nos gráficos com 5 clusters, observou-se uma separação relativamente clara entre alguns grupos, especialmente em combinações como mode vs instrumentalness e acousticness vs instrumentalness. Essas combinações apresentam variabilidade suficiente para permitir uma distinção significativa entre os pontos, o que também é consistente com os altos índices de silhueta obtidos previamente.

Por outro lado, ao aumentar o número de clusters para o total de gêneros presentes, a separação visual entre os grupos tornou-se mais fragmentada. Entretanto, em certos casos, como na combinação danceability vs energy, o uso dos 59 clusters permitiu uma análise mais refinada. Isso se deve ao fato de que esses dois atributos são contínuos e apresentam uma grande dispersão de valores, o que favorece uma subdivisão mais detalhada dos dados e, consequentemente, pode capturar nuances mais sutis entre estilos musicais diferentes. Com muitos dados disponíveis nesses eixos, a granularidade oferecida por um número maior de clusters não resulta em sobreajuste, mas sim em maior resolução na análise.

Em contraste, na combinação key vs mode, que utiliza atributos mais discretos e com menos variabilidade (como mode, que é binário, e key, que assume apenas 12 valores inteiros), observou-se que o agrupamento com apenas 5 clusters resultou em uma visualização mais limpa e interpretável. Isso ocorre porque um número muito alto de clusters, como 59, em um espaço com baixa dimensionalidade e baixa dispersão, gera sobrefragmentação — criando grupos artificiais que não refletem variações reais no conjunto de dados.

O seguinte código apresenta a implementação da interface interativa utilizando o pacote ipywidgets, com o objetivo de recomendar as músicas com base em filtros ajustáveis definidos pelo usuário. A interface é composta por diversos sliders e campos de entrada, que permitem especificar faixas desejadas para características musicais como instrumentalness, energia, popularidade, acousticness, duração e também um campo de texto para o gênero musical. Ao clicar no botão "Recomendar músicas", a função recomendar_musicas_avancado é executada, filtrando um DataFrame de músicas (df) conforme os critérios estabelecidos e retornando as músicas mais populares que atendem aos parâmetros escolhidos, até o número máximo definido pelo usuário. O resultado é exibido diretamente na célula, com colunas relevantes como nome da música, artista, gênero e características sonoras. Essa abordagem permite uma análise exploratória personalizada e dinâmica do conjunto de dados musicais, facilitando a descoberta de músicas que correspondam ao gosto do usuário.

```python
import ipywidgets as widgets
from IPython.display import display

# Sliders para entrada do usuário
genero_text = widgets.Text(value='', description='Gênero (ex: pop)', placeholder='Deixe vazio para qualquer')
instrumentalness_min_slider = widgets.FloatSlider(value=0.0, min=0, max=1, step=0.05, description='Instrum. min')
instrumentalness_max_slider = widgets.FloatSlider(value=1.0, min=0, max=1, step=0.05, description='Instrum. max')
energy_min_slider = widgets.FloatSlider(value=0.7, min=0, max=1, step=0.05, description='Energia min')
energy_max_slider = widgets.FloatSlider(value=1.0, min=0, max=1, step=0.05, description='Energia max')
popularity_min_slider = widgets.IntSlider(value=0, min=0, max=100, step=1, description='Popularidade min')
popularity_max_slider = widgets.IntSlider(value=100, min=0, max=100, step=1, description='Popularidade max')
acousticness_min_slider = widgets.FloatSlider(value=0.0, min=0, max=1, step=0.05, description='Acústica min')
acousticness_max_slider = widgets.FloatSlider(value=1.0, min=0, max=1, step=0.05, description='Acústica max')
duration_slider = widgets.IntSlider(value=300, min=30, max=600, step=10, description='Duração (s)')
top_n_slider = widgets.IntSlider(value=10, min=1, max=30, step=1, description='Qtd. músicas')

# Botão
botao = widgets.Button(description='Recomendar músicas')

# Função avançada de recomendação
def recomendar_musicas_avancado(
    df,
    genero=None,
    instrumentalness_min=0.0,
    instrumentalness_max=1.0,
    energy_min=0.0,
    energy_max=1.0,
    popularity_min=0,
    popularity_max=100,
    acousticness_min=0.0,
    acousticness_max=1.0,
    duration_max=None,
    top_n=30
):
    df_filtrado = df.copy()

    if genero and genero.strip() != '':
        df_filtrado = df_filtrado[df_filtrado['genre'].str.contains(genero, case=False, na=False)]

    df_filtrado = df_filtrado[
        (df_filtrado['instrumentalness'] >= instrumentalness_min) &
        (df_filtrado['instrumentalness'] <= instrumentalness_max) &
        (df_filtrado['energy'] >= energy_min) &
        (df_filtrado['energy'] <= energy_max) &
        (df_filtrado['popularity'] >= popularity_min) &
        (df_filtrado['popularity'] <= popularity_max) &
        (df_filtrado['acousticness'] >= acousticness_min) &
        (df_filtrado['acousticness'] <= acousticness_max)
    ]

    if duration_max is not None:
        df_filtrado = df_filtrado[df_filtrado['duration_ms'] <= duration_max * 1000]

    df_filtrado = df_filtrado.sort_values(by='popularity', ascending=False)

    return df_filtrado[['song', 'artist', 'genre', 'instrumentalness', 'energy', 'popularity', 'acousticness', 'duration_ms']].head(top_n)

# Função chamada ao clicar no botão
def on_click(b):
    resultados = recomendar_musicas_avancado(
        df,
        genero=genero_text.value,
        instrumentalness_min=instrumentalness_min_slider.value,
        instrumentalness_max=instrumentalness_max_slider.value,
        energy_min=energy_min_slider.value,
        energy_max=energy_max_slider.value,
        popularity_min=popularity_min_slider.value,
        popularity_max=popularity_max_slider.value,
        acousticness_min=acousticness_min_slider.value,
        acousticness_max=acousticness_max_slider.value,
        duration_max=duration_slider.value,
        top_n=top_n_slider.value
    )
    display(resultados)

# Conectar botão à função
botao.on_click(on_click)

# Mostrar interface
display(
    genero_text,
    instrumentalness_min_slider,
    instrumentalness_max_slider,
    energy_min_slider,
    energy_max_slider,
    popularity_min_slider,
    popularity_max_slider,
    acousticness_min_slider,
    acousticness_max_slider,
    duration_slider,
    top_n_slider,
    botao
)
```
